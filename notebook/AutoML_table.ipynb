### デプロイコスト

AutoMLでonline predictまでやろうとすると、デプロイする必要がある。  
一方で、Batch Predictなら必要はないようである。  

GiB（ギビバイト）当たり1時間0.005ドル。  
[1ギビバイト ＝ 1,024メガバイト](https://wa3.i-3-i.info/word11062.html)。  
レイテンシーを下げるために9つのマシンにデプロイしているため価格は9倍となる。

例えば、10GBのモデルを3時間デプロイしておくと、以下のような計算になる。
```
0.005 * 10 * 3 * 9
```
